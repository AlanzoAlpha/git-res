# Top ten movies that got rated the most

```python
Top_rated = df.groupby('title').size().sort_values(ascending=False)
Top_rated.head(10)
```

# Top ten movies that received highest number of "5" ratings

```python
dfs=df
dfs = dfs[(dfs['rating']==5)]
Five_star = dfs.groupby('title')['rating'].count().sort_values(ascending = False)
Five_star.head(10)
```

# Ten most rated movie by students

```python

dfst= df[(df['occupation']=='student')]
dfst=dfst.groupby('title').count().sort_values(['rating'],ascending=False)
dfst[['rating']].head(10)
```

# Top ten most rated movie between ages 20 and 25

```python
dfa = df[(df['age']>=20) & (df['age']<=25)]
dfa =dfa.groupby('title').count().sort_values(['rating'],ascending=False)
dfa[['rating']].head(10)
```

# Top ten most rated by zipcode

```python
dfz = df.groupby('zip_code').count().sort_values(['rating'],ascending=False)
dfz[['rating']].head(10)
```

# Top ten most rated movies after 1960

```python
ratings['release_date'] = ratings['unix_timestamp'].apply(lambda x: datetime.fromtimestamp(x))
ratings_under = ratings[ratings['release_date'] > datetime(1960,12,31)]
movie_ids= ratings_under['movie_id']
movies_under = movies[movies['movie_id'].isin(movie_ids)]
movies_under['title'].head(10)
```
