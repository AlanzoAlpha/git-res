# Write a Pandas program to create a Pivot table with the following indexes from the data set of titanic.csv a. Sex - age b. Sex - Pclass c. Sex â€“ survived

```python
sa= pd.pivot_table(titanic, index = ["Sex","Age"], aggfunc=np.sum)
sa.head(10).append(sa.tail(10))
```

```python
sp = pd.pivot_table(titanic, index = ["Sex","Pclass"], aggfunc=np.sum)
sp.head(10).append(sp.tail(10))
```

```python
ss = pd.pivot_table(titanic, index = ["Sex","Survived"], aggfunc=np.sum)
ss.head(10).append(ss.tail(10))
```

# Write a Pandas program to partition each of the passengers into four categories based on their age

```python
labels = ['Kids','Teens','Adults','Elders']
titanic['AgeCat'] = pd.cut(titanic['Age'], bins =[0, 12, 20, 60, 100], labels = labels , right = False)
a = titanic.groupby('AgeCat')['Name'].count()
a.head(10)

```

# Write a Pandas program to group children  < =18 age.

```python
ct = titanic[titanic['Age'] <= 18].groupby('Age')['Name'].count()
ct.head(24)
```
