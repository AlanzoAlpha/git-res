```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Draw a pie chart to show any five first world dept in billion of dollars

```python
First_world_dept = [22, 12, 7, 3, 2]
country_lb = ['USA', 'Japan', 'China', 'Italy','UK']
plt.pie(First_world_dept, labels = country_lb  )
```

```python
plt.pie(First_world_dept, labels = country_lb , shadow = True, autopct = '%1.0f%%', radius = 10)
plt.axis('equal')
plt.show()
```

```python
plt.pie(First_world_dept, labels = country_lb , shadow = True, autopct = '%1.0f%%', radius = 10, explode = [0,0,0,0,4])
plt.axis('equal')
plt.show()
```

```python
First_world_dept = np.array([22, 12, 7, 3, 2])
country_lb = ['USA', 'Japan', 'China', 'Italy','UK']
colors = ['red','white' ,'gold','green', 'Blue']
def absolute_value(val):
    a  = np.round(val/100.*First_world_dept.sum(), 0)
    return a
plt.pie(First_world_dept, labels = country_lb , colors = colors, autopct = absolute_value,shadow = True, radius = 10, explode=[0,0,0,0,4],counterclock = True, startangle = 90)
plt.axis('equal')
plt.show()
```
