```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Create a bar chart to show 5 top tech companies revenue

```python
company=['GOOGLE','AMAZON','MICROSOFT','FACEBOOK','APPLE']
revenue=[162,233,126,71,260]
```

```python
pos = np.arange(len(company))
```

```python
plt.bar(pos,revenue, label="revenue")

plt.xticks(pos,company)
plt.ylabel("revenue(B$)")
plt.title('Top technology Stocks')
plt.legend()
```

# Add the companies profit on the chart. 

```python
profit = [50,74,46,20,57]
plt.bar(pos-0.2,revenue, width=0.4, label="revenue")
plt.bar(pos+0.2,profit, width=0.4,label="profit")

plt.xticks(pos,company)
plt.ylabel("revenue(B$)")
plt.title('Top technology Stocks')
plt.legend()
```

```python
profit = [50,74,46,20,57]
plt.barh(pos-0.2,revenue, label="revenue")
plt.barh(pos+0.2,profit,label="profit")

plt.yticks(pos,company)
plt.xlabel("revenue & profit(B$)")
plt.xlabel("revenue & profit(B$)")
plt.title('Top technology Stocks')
plt.legend()
```
