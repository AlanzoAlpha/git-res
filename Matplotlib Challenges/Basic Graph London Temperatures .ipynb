```python
import matplotlib.pyplot as plt
from datetime import date
%matplotlib inline
```

```python
today = date.today()
x = today.strftime("%B %d, %Y")
print(f'Today is {x}. \nHere is a list of the high,low and mean temperature in london for the next 7 days, starting today. ')
days=[1,2,3,4,5,6,7]
max_t=[21,21,22,21,22,22,20]
min_t=[15,17,12,11,12,13,13]
avg_t=[18,19,17,16,17,18,17]
print(f'Max temperatures are {max_t} \nMin Temperatures are {min_t} \nMean temperature are {avg_t}')
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

# Inserting labels

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

# Add lendgend

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
```

# Moving lengend around and increasing font size.

```python

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```
